```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('placement[1].csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
plt.boxplot(dataset['package'])
```

```python
x=dataset['cgpa'].values.reshape(-1,1)
y=dataset['package'].values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,Y_train)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
r_squared = lin_reg.score(X_train, Y_train)
```

```python
print(r_squared)
```

```python
plt.scatter(x = X_train, y = Y_train, color = 'red')
plt.plot(X_train, lin_reg.predict(X_train), color = 'blue')
plt.title('Package Vs CGPA (Training Data)')
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
y_pred = lin_reg.predict(X_test)
df = pd.DataFrame({'Actual': Y_test[:,0], 'Predicted': y_pred[:,0]})
df
```

```python
plt.scatter(x = X_test, y = Y_test, color = 'red')
plt.plot(X_train, lin_reg.predict(X_train), color = 'blue')
plt.title('Package Vs CGPA (Test Data)')
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
cgpa = [[8.5]]
pred = lin_reg.predict(cgpa)
print("CGPA of student is: {}".format(cgpa[0]))
print("Package offered: {}".format(pred[0]))
```
