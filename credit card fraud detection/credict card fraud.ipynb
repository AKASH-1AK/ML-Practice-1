## importing the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
## Seperating the data for Analysing
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

## Under Sampling

```python
# building a sample dataset containing of ligit transaction and fraudlent transaction

#number of fradulent transaction --> 492
legit_sample = legit.sample(n=492)
```

```python
# concatinating two data frames
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

## Data training and SPlitting

```python
x=new_data.drop(columns='Class',axis=1)
y=new_data['Class']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

## Logistic Regression

```python
model = LogisticRegression()
```

```python
#Training the logistic regression with train data.
model.fit(x_train,y_train)
```

## model evaluation

```python
## accuracy score on train data:

x_train_pred = model.predict(x_train)
train_acc = accuracy_score(x_train_pred,y_train)
print(train_acc)
```

```python
## acc on test data:
x_test_pred = model.predict(x_test)
test_acc = accuracy_score(x_test_pred,y_test)
print(test_acc)
```
